```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('cat.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('do-not-copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
print("Image 1 Shape: ",img1.shape)
print("Image 2 Shape: ",img2.shape)
```

## Blending Image of Same Size

```python
img1_resize = cv2.resize(img1, (300,300))
img2_resize = cv2.resize(img2, (300,300))
```

```python
plt.imshow(img1_resize)
```

```python
plt.imshow(img2_resize)
```

```python
blended_img = cv2.addWeighted(src1 = img1_resize, alpha=0.9, src2 = img2_resize, beta=0.2, gamma=0)
plt.imshow(blended_img)
```

## Overlay Small Image On the top of large image (No Blended)
## 1)  Numpy Re-Assignment Operation

```python
img1 = cv2.imread('cat.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('do-not-copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2_resize = cv2.resize(img2, (200,200))
plt.imshow(img2_resize)
```

```python
plt.imshow(img1)
```

```python
large_img = img1
small_img = img2_resize
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0] 
```

```python
large_img[y_offset:y_end, x_offset:x_end]= small_img
```

```python
plt.imshow(large_img)
```
