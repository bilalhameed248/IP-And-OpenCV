```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
blank_img=np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
print(blank_img.shape)
print(blank_img.dtype)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(300,100),pt2=(500,300),color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(100,300),pt2=(200,400),color=(255,0,0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,100), radius=(50), color=(0,0,255), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(400,400), radius=(50), color=(0,0,255), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(img=blank_img, pt1=(100,300),pt2=(200,400), color=(0,0,200), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.line(img=blank_img, pt1=(0,0),pt2=(512,512), color=(0,0,200), thickness=10)
plt.imshow(blank_img)
```

## PART 2

Writing text on images and Polygons 

```python
blank_img=np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text = "Hello", org=(10,500), fontFace=font, fontScale=3, color=(255,0,0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img=np.zeros(shape=(512,512,3), dtype=np.int32)
```

```python
vertices=np.array([[100,300],[200,200],[400,300],[200,400]], dtype=np.int32 )
```

```python
vertices
```

```python
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
```

```python
pts
```

```python
pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed=True, color=(255.0,0), thickness=10)
plt.imshow(blank_img)
```
