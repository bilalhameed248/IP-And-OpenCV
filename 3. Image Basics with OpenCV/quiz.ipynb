Open the image from folder and display it in notebook and correct the RGB Order

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img=cv2.imread('colorpic.jpg')
fix_image=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

Flip the image into upside down and display it into notebook

```python
flip_img=cv2.flip(fix_image, 0)
plt.imshow(flip_img)
```

Draw red Rectangle on image

```python
img1=cv2.rectangle(fix_image, pt1=(400,200),pt2=(600,400),color=(255,0,0), thickness=10)
plt.imshow(img1)
```

```python
pt1 = (150, 200)
pt2 = (100, 400)
pt3 = (200, 600)
triangle_cnt = np.array( [pt1, pt2, pt3] )
triangle_cnt
```

```python
tri_img = cv2.drawContours(fix_image, [triangle_cnt], 0, (0,255,0), thickness=10)
plt.imshow(tri_img)
```

```python
tri_img = cv2.drawContours(fix_image, [triangle_cnt], 0, (0,255,0), -1)
plt.imshow(tri_img)
```

```python
import cv2
import numpy as np

####################################
############ Functions #############
####################################

def draw_circle(event, x, y, flag, param):
    if event==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y),100,(0,255,0),-1)
        
    elif event==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x,y),100,(0,0,255),thickness=10)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing', draw_circle)



###################################
### Showing Image with Open CV ####
###################################
img=np.zeros((512,512,3))
while True:
    cv2.imshow('my_drawing', img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
