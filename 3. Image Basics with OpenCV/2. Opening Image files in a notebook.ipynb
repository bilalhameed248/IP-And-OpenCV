```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img=cv2.imread('colorpic.jpg')
```

```python
plt.imshow(img)
```

```python
print(type(img))
print(img.shape)
```

```python
# MATPLOTLIB --> RED GREEN BLUE
# OPEN CV  --> BLUE GREEN RED
```

```python
fix_image=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_image)
```

## Image Resizing

```python
resize_img=cv2.resize(fix_image, (500,200))
plt.imshow(resize_img)
```

```python
w_raito=0.5
h_raito=0.2
```

```python
resize_img_by_raito=cv2.resize(fix_image,(0,0), fix_image,w_raito, h_raito)
plt.imshow(resize_img_by_raito)
```

## Flip Image

```python
flip_image=cv2.flip(fix_image,0)
plt.imshow(flip_image)
```

```python
flip_image=cv2.flip(fix_image,1)
plt.imshow(flip_image)
```

```python
flip_image=cv2.flip(fix_image,1)
# plt.figure(figsize=(800,800))
plt.imshow(flip_image)
# cv2.imwrite('flip_color_image.jpg',flip_image)
```
