```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('colorpic.jpg')
```

```python
type(pic)
```

```python
pic_array=np.asarray(pic)
```

```python
pic_array.shape
```

```python
plt.imshow(pic_array)
```

```python
pic_red=pic_array.copy()
```

```python
# R, G, B (0, 1, 2)
pic_red[:,:,1]
```

```python
# Red Channel Value (0-255) 255 mean full pure red and 0 mean pure black
plt.imshow(pic_red[:,:,0], cmap='gray')
```

```python
# Green Channel Value (0-225)
plt.imshow(pic_red[:,:,1], cmap='gray')
```

```python
# Blue Channel Value (0-225)
plt.imshow(pic_red[:,:,2], cmap='gray')
```
